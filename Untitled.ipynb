## Utils

```python

toOrderedPair = lambda X,Y: np.array(list(map(lambda xy: np.array(xy),list(product(X.flatten(),Y.flatten())))))
def scatterPlot2d(X,**kwargs):
    _,ax   = plt.subplots(figsize=(7,7))
    colors = kwargs.get('colors',['#CDDC39','#673AB7',"#E91E63","#03A9F4"])
    alpha  = kwargs.get('alpha',1)
    for i,wi in enumerate(X):
        ax.scatter(wi[:,0],wi[:,1],color=colors[i],alpha=alpha)
    return ax
def getValuesByClass(X,labels,**kwargs):
    n_classes = kwargs.get('n_classes',2)
    classes   = range(n_classes)
    indices   = map(lambda label: np.argwhere(labels==label).flatten(),classes)
    values    = list(map(lambda index:X[index],list(indices)))
    return values
def zipWithLabel(X,labels):
    result=dict()
    for xi,yi in zip(X,y):
        if not yi in result:
            result[yi]=[]
        result[yi].append(xi)
    return dict((k, np.array(v)) for k, v in result.items())
```
