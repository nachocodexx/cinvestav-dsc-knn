## Utils

```python

# np.array(list(map(lambda xy: np.array(xy),list(product(X.flatten(),Y.flatten())))))
def scatterPlot2d(X,**kwargs):
    _,ax   = plt.subplots(figsize=(7,7))
    colors = kwargs.get('colors',COLORS)
    alpha  = kwargs.get('alpha',1)
    for i,wi in enumerate(X):
        ax.scatter(wi[:,0],wi[:,1],color=colors[i],alpha=alpha)
    return ax
def getValuesByClass(X,labels,**kwargs):
    n_classes = kwargs.get('n_classes',2)
    classes   = range(n_classes)
    indices   = map(lambda label: np.argwhere(labels==label).flatten(),classes)
    values    = list(map(lambda index:X[index],list(indices)))
    return values
def zipWithLabel(X,labels):
    result=dict()
    for xi,yi in zip(X,y):
        if not yi in result:
            result[yi]=[]
        result[yi].append(xi)
    return dict((k, np.array(v)) for k, v in result.items())
```

```python
def toOrderedPair(X,Y):
    xShape= X.shape
    yShape= Y.shape
#     print(xShape,yShape)
    return np.dstack([X,Y]).reshape(xShape[0]*yShape[0],2)
# toOrderedPair(_X,Y).shape
# np.dstack([_X,Y]).reshape(100,2)
```

```python
def knn2DPlot(**kwargs):
    Zs,XY     = kwargs.get("Zs",([],[]))
    values = kwargs.get("X",[])
#     print(XY.shape)
#     XY     = kwargs.get('XY',[])
    title  = kwargs.get('title','Default')
    _,ax   = plt.subplots(figsize=(5,5))
    for i,indices in enumerate(Zs):
        Xs = XY[indices]
        ax.scatter(Xs[:,0],Xs[:,1],color=COLORS[i],alpha=.5)
    for i,xi in enumerate(values):
        ax.scatter(xi[:,0],xi[:,1],
                   color=COLORS[i],
                   alpha=1,
                   edgecolors="white",
                   label="$\omega_{}$".format(i),
                   linewidths=1)
    ax.set_title(title)
    ax.set_xlabel("$x_1$")
    ax.set_ylabel("$x_2$")
    ax.legend(framealpha=1)
```

```python
def getDecisionBoundary(**kwargs):
    xLim    = kwargs.get('xLim',1.5)
    step    = kwargs.get('step',.03)
    knn     = kwargs.get('knn',None)
    xyLen   = np.arange(-xLim, xLim, step)
    _X,Y    = np.meshgrid(xyLen,xyLen)
    XY      = toOrderedPair(_X,Y) 
    Z       = knn.predict(XY)
    classes = np.unique(Z)
    result  = []
    for i in classes:
        result.append(np.argwhere(Z==i).flatten())
    return result,XY
```
